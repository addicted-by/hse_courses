```python
df = pd.read_excel('WaitData.Published.xlsx', sheet_name='F3')
```

```python
df = df[~df['DayOfYear'].isna()]
```

```python
features = list(set(df.columns) - {'Wait', 'x_ArrivalDTTM', 'x_ScheduledDTTM', 'x_BeginDTTM'})
X = df[features]
```

```python
y = df['Wait'].values
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X, y)
Ypred = model.predict(X) # use trained regression model to predict
r = y-Ypred # compute prediction error (residual)
e = abs(r).mean() # compute model error
```

```python
def get_error(model, X, y):
    model.fit(X, y)
    Ypred = model.predict(X)
    r = y-Ypred
    return abs(r).mean()
```

```python
min_error, best_feature = None, None
for feature in X.columns:
    error = get_error(model, X[[feature]], y)
    print(feature, error)
    if min_error is None or min_error > error:
        min_error = error
        best_feature = feature
```

```python
def iterative_feature_selection(model, X, y):
    selected_features = []
    current_min_error = None
    for i in range(15):
        min_error, best_feature = None, None
        for feature in set(X.columns) - set(selected_features):
            error = get_error(model, X[selected_features + [feature]], y)
            if min_error is None or min_error > error:
                min_error = error
                best_feature = feature
        current_min_error = min_error
        selected_features.append(best_feature)
        print(i, current_min_error)
```

```python
iterative_feature_selection(model, X, y)
```

```python
def brute_feature_selection(model, X, y):
    min_error, best_features = None, None
    for first_feature_idx in range(len(X.columns) - 2):
        for second_feature_idx in range(first_feature_idx + 1, len(X.columns) - 1):
            for third_feature_idx in range(second_feature_idx + 1, len(X.columns)):
                features = [X.columns[first_feature_idx], X.columns[second_feature_idx], X.columns[third_feature_idx]]
                error = get_error(model, X[features], y)
                if min_error is None or min_error > error:
                    min_error = error
                    best_features = features
    return min_error, best_features
```

```python
brute_feature_selection(model, X, y)
```
